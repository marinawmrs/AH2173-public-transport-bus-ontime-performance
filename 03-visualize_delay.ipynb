# SETUP

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# read trip and weather data 
control = pd.read_csv('./weather_delay_CONTROL.csv')
snow = pd.read_csv('./weather_delay_SNOW.csv')
combined = pd.concat([control, snow], ignore_index=True).sort_values("date")
```

# Visualization

```python
def plot_metric(df, metric):
  # Plot bar chart of DRI per Trip by date
  plt.figure(figsize=(14, 6))
  sns.barplot(data=df, x='date', y=metric)

  # Formatting
  plt.title(metric + " by Date")
  plt.xlabel("Date")
  plt.ylabel(metric)
  plt.xticks(rotation=45, ha='right')
  plt.legend(title="Snow Day", loc="upper right")
  plt.tight_layout()
  plt.show()
```

```python
plot_metric(combined, "mean_arrival_delay")
```

```python
delay_by_ground_condition = pd.DataFrame(combined.groupby("ground_expl")["date"].count())
plt.figure(figsize=(5, 6))
sns.barplot(data=delay_by_ground_condition, x='ground_expl', y='date')

# Formatting
plt.xlabel("Ground condition")
plt.ylabel("Count")
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
